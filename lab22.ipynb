```python
# Install dependencies as needed:
# pip install kagglehub[pandas-datasets]
import kagglehub
from kagglehub import KaggleDatasetAdapter

# Set the path to the file you'd like to load
file_path = "ecommerce_sales.csv"

# Load the latest version
df = kagglehub.load_dataset(
  KaggleDatasetAdapter.PANDAS,
  "nmnbabbar/ecommerce-sales",
  file_path,
  # Provide any additional arguments like
  # sql_query or pandas_kwargs. See the
  # documenation for more information:
  # https://github.com/Kaggle/kagglehub/blob/main/README.md#kaggledatasetadapterpandas
)

print("First 5 records:", df.head())
```

```python
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
monthly_sales = df.resample('M', on='Date')['Revenue'].sum()

monthly_sales.plot(figsize=(8,5), title='Monthly Revenue')
plt.ylabel("Revenue")
plt.show()
```

```python

weekly_sales = df.resample('W', on='Date')['Revenue'].sum()

weekly_sales.plot(figsize=(8,5), title='Weekly Revenue')
plt.ylabel("Revenue")
plt.show()
```

```python
daily_sales = df.resample('D', on='Date')['Revenue'].sum()

daily_sales.plot(figsize=(10,5), title='Daily Revenue')
plt.ylabel("Revenue")
plt.show()
```

```python
df.corr(numeric_only=True)
```

```python
import seaborn as sns

plt.figure(figsize=(6,4))

sns.heatmap(df.corr(numeric_only=True),
            annot=True,
            cmap="coolwarm")

plt.title("Correlation Heatmap - Sales Data")
plt.show()
```

Data drilling was performed to analyze revenue at monthly, weekly, and daily levels.
• Monthly revenue plots show how sales trends change over time.
• Weekly and daily analysis provides more detailed insights into sales performance.
• The correlation heatmap shows relationships between numerical variables.
• Since revenue is the primary numeric variable, correlation results are limited.

```python
# Install dependencies as needed:
# pip install kagglehub[pandas-datasets]
import kagglehub
from kagglehub import KaggleDatasetAdapter

# Set the path to the file you'd like to load
file_path = "Listings.csv"

# Load the latest version
df = kagglehub.dataset_load(
  KaggleDatasetAdapter.PANDAS,
  "ulrikthygepedersen/airbnb-listings",
  file_path,
  pandas_kwargs={'encoding': 'latin1'}
  # Provide any additional arguments like
  # sql_query or pandas_kwargs. See the
  # documenation for more information:
  # https://github.com/Kaggle/kagglehub/blob/main/README.md#kaggledatasetadapterpandas
)

print("First 5 records:", df.head())
```

```python
df.columns
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

sns.boxplot(x=df['price'])

plt.title("Outlier Detection in Airbnb Prices")
plt.show()
```

```python
import numpy as np
from scipy import stats

z = np.abs(stats.zscore(df['price'].dropna()))

outliers = df[z > 3]

print("Number of Outliers:", len(outliers))

outliers.head()
```

```python
plt.figure(figsize=(8,5))

sns.barplot(x='room_type', y='price', data=df)

plt.title("Average Price by Room Type")
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(x='room_type', y='price', data=df)

plt.title("Price Distribution by Room Type")
plt.show()
```


A boxplot was used to detect outliers in Airbnb listing prices.
• Some listings have unusually high prices compared to the majority of listings.
• Z-score method was used to identify extreme values.
• Barplots show the relationship between room type and price.
• Entire homes generally have higher average prices than private rooms.

```python
# Install dependencies as needed:
# pip install kagglehub[pandas-datasets]
import kagglehub
from kagglehub import KaggleDatasetAdapter

# Set the path to the file you'd like to load
file_path = "owid-covid-data.csv"

# Load the latest version
df = kagglehub.load_dataset(
  KaggleDatasetAdapter.PANDAS,
  "abubakaryagob/covid-19",
  file_path,
  # Provide any additional arguments like
  # sql_query or pandas_kwargs. See the
  # documenation for more information:
  # https://github.com/Kaggle/kagglehub/blob/main/README.md#kaggledatasetadapterpandas
)

print("First 5 records:", df.head())
```

```python
print("Missing Values in Each Column:")
print(df.isnull().sum())
```

```python
missing_percent = (df.isnull().sum() / len(df)) * 100
print(missing_percent.sort_values(ascending=False))
```

```python
df.fillna(method='ffill', inplace=True)
```

```python
print("Missing Values After Cleaning:")
print(df.isnull().sum())
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
pakistan = df[df['location'] == 'Pakistan']
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))

sns.lineplot(x='date', y='new_cases', data=pakistan)

plt.title("COVID-19 Daily Cases in Pakistan")
plt.xlabel("Date")
plt.ylabel("New Cases")
plt.show()
```

```python
import numpy as np
from scipy import stats

z = np.abs(stats.zscore(pakistan['new_cases'].dropna()))

anomalies = pakistan[z > 3]

print("Number of Anomalies:", len(anomalies))

anomalies.head()
```

• The dataset contains global COVID-19 statistics including cases, deaths, and health indicators.
• Missing values were detected using isnull() and handled using forward fill.
• This method replaced missing values with the previous available data.
• A line plot was used to visualize daily COVID-19 cases.
• Sudden spikes in daily cases indicate anomalies in the dataset.
