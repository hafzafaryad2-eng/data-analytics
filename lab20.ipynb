```python
# 1. Mean High, Mean Low, Average Daily Range
# =============================
mean_high = df_weather["High_F"].mean()
mean_low = df_weather["Low_F"].mean()
avg_range = mean_high - mean_low
print(f"Mean High Temp: {mean_high:.2f}°F")
print(f"Mean Low Temp: {mean_low:.2f}°F")
print(f"Average Daily Temp Range: {avg_range:.2f}°F")
print("-"*50)
```

```python
# 2. MAX and MIN Summary
# =============================
summary = {
    "High_F": {"MAX": df_weather["High_F"].max(), "MIN": df_weather["High_F"].min()},
    "Wind_mph": {"MAX": df_weather["Wind_mph"].max(), "MIN": df_weather["Wind_mph"].min()},
    "Precip_in": {"MAX": df_weather["Precip_in"].max(), "MIN": df_weather["Precip_in"].min()},
    "AQI": {"MAX": df_weather["AQI"].max(), "MIN": df_weather["AQI"].min()}
}
print("Summary of MAX and MIN values:")
for col, vals in summary.items():
    print(f"{col}: MAX = {vals['MAX']}, MIN = {vals['MIN']}")
print("-"*50)
```

```python
# 3. STD DEV & Elevated-Pollution Days
# =============================
aqi_mean = df_weather["AQI"].mean()
aqi_std = df_weather["AQI"].std()
elevated_days = df_weather[df_weather["AQI"] > (aqi_mean + aqi_std)]["Date"].tolist()
print(f"AQI Mean: {aqi_mean:.2f}, STD DEV: {aqi_std:.2f}")
print("Elevated-Pollution Days (AQI > mean + 1 std dev):", elevated_days)
print("-"*50)
```

```python
# 4. Weekend vs Weekday Analysis
# =============================
weekend = df_weather[df_weather["Day_Type"]=="Weekend"]
weekday = df_weather[df_weather["Day_Type"]=="Weekday"]

def summarize(df):
    return {
        "Mean_High": df["High_F"].mean(),
        "Mean_Humidity": df["Humidity"].mean(),
        "Total_Precip": df["Precip_in"].sum()
    }

weekend_stats = summarize(weekend)
weekday_stats = summarize(weekday)
print("Weekend Stats:", weekend_stats)
print("Weekday Stats:", weekday_stats)
print("-"*50)
```

```python
# 5. Mode of Day Type
# =============================
from statistics import mode
day_type_mode = mode(df_weather["Day_Type"])
print("Mode of Day Type:", day_type_mode)
print("Observation: Most readings are from weekdays. Mode less useful because only 2 categories exist.")
print("-"*50)
```

```python
# 6. Median and Mean of Precipitation
# =============================
median_precip = df_weather["Precip_in"].median()
mean_precip = df_weather["Precip_in"].mean()
print(f"Median Precipitation: {median_precip:.2f} in")
print(f"Mean Precipitation: {mean_precip:.2f} in")
print("Observation: Median is more representative; Mean is skewed by heavy rain days.")
print("-"*50)
```

```python
# 7. Compare Wind Speed and AQI
# =============================
# Sort by Wind Speed descending
sorted_df = df_weather.sort_values(by="Wind_mph", ascending=False)
for i, row in sorted_df.iterrows():
    print(f"Date: {row['Date']}, Wind: {row['Wind_mph']} mph, AQI: {row['AQI']}")
print("Observation: High wind days (03/13, 03/12, etc.) do not always have low AQI. Correlation seems weak but can reduce pollution slightly on some days.")
```

Exercise 5A — GreenCart Insights

1. Revenue by Product Category
From the aggregation in Dataset A, Beverages and Snacks generated the highest total revenue (e.g., Sparkling Water 12pk: $746.17, Mixed Nuts: $455.52), while Produce had the lowest total revenue overall (e.g., Baby Kale Mix: $123.69, Avocados: $141.55). A store manager could use this insight to promote underperforming categories through discounts or bundle offers, while maintaining marketing focus on high-revenue categories to maximize profits. Targeted promotions could increase sales for low-revenue items without compromising the top sellers.

2. Mean vs Median Revenue and Skew
The mean revenue ($314.77) is higher than the median ($221.73), indicating right skew caused by very high-revenue orders such as Sparkling Water ($746.17) and Granola Bar 6pk ($439.12). These are outliers in the dataset. In a cleaner analysis, we might report median revenue for typical performance, while analyzing outliers separately to understand their effect on overall revenue trends.

3. Region Targeting for Promotions
Examining the average revenue per region, West and North regions show higher average order values (e.g., West: Sparkling Water $746.17; North: Green Tea $395.28). Therefore, I recommend targeting West for promotional campaigns, as it combines high revenue potential with multiple orders above $300, ensuring campaigns reach profitable customers. This evidence-based approach maximizes the likelihood of ROI on marketing efforts.

3. High Mean vs Low STD DEV Team Preference
A high mean score with high standard deviation indicates excellent performers but inconsistency, while a lower mean with low std dev shows consistent but moderate performance. For a manager, I would prefer lower mean with low standard deviation, because consistency ensures reliable performance across projects. High variance can be risky if top performers are unavailable or underperform unexpectedly, even if average scores are higher.
